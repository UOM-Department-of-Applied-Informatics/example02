## Regex

Η standard βιβλιοθήκη κανονικών εκφράσεων [`RegEx`](https://docs.python.org/3/howto/regex.html) (δεν χρειάζεται ξεχωριστή εγκατάσταση) επιτρέπει την αναζήτηση κειμένου με ευέλικτα κριτήρια.
Το παρακάτω πρόγραμμα για παράδειγμα ελέγχει αν η συμβολοσειρά ξεκινά με το "The" και τελειώνει με το "Spain".
```Python
import re
#Check if the string starts with "The" and ends with "Spain":
txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)
if x:
  print("YES! We have a match!")
else:
  print("No match")
```
Εμφανίζει...
```Console
YES! We have a match!
```


Το παρακάτω πρόγραμμα επιστρέφει μια λίστα με όλα τα "ai" σε μια συμβολοσειρά.
```Python
import re
#Return a list containing every occurrence of "ai":
txt = "The rain in Spain"
x = re.findall("ai", txt)
print(x)
```
Εμφανίζει...
```Console
['ai', 'ai']
```


Το παρακάτω πρόγραμμα αντικαθιστά όλα τα κενά με το "9" σε μια συμβολοσειρά.
```Python
import re
#Replace all white-space characters with the digit "9":
txt = "The rain in Spain"

print(x)
```
Εμφανίζει...

## Βιβλιοθήκη JSON

Η βιβλιοθήκη `json` χρησιμοποιείται για να μετατρέψει δεδομένα μεταξύ Python και μορφής JSON.

````python
# Δοκίμασε το παράδειγμα μετατροπής Python σε JSON
import json

data = {'name': 'Anna', 'age': 23, 'city': 'Athens'}
json_str = json.dumps(data)
print(json_str)

# Μετατροπή JSON σε Python
python_data = json.loads(json_str)
print(python_data)
````

## Διαχείριση Εξαιρέσεων (try-except)

Ο μηχανισμός try-except χρησιμοποιείται για να διαχειριστούμε λάθη που μπορεί να προκύψουν κατά την εκτέλεση ενός προγράμματος.

````python
# Δοκίμασε το παρακάτω παράδειγμα try-except

try:
    num = int(input('Δώσε έναν αριθμό: '))
    print('Έδωσες τον αριθμό:', num)
except ValueError:
    print('Αυτό που έδωσες δεν ήταν αριθμός!')
````

### Πρόγραμμα με ορίσματα γραμμής εντολών:

Μην δοκιμάσετε να το εκτελέσετε σε code block. Φτιάξτε ξεχωριστό αρχείο "**add_numbers.py**"

```Python
import sys

# Έλεγχος πλήθους ορισμάτων
if len(sys.argv) != 3:
    print("Λάθος αριθμός ορισμάτων. Χρήση: python add_numbers.py <num1> <num2>")
    sys.exit(1)

# Προσπάθεια μετατροπής των ορισμάτων σε αριθμούς
try:
    num1 = float(sys.argv[1])
    num2 = float(sys.argv[2])
    result = num1 + num2
    print(f"Το άθροισμα των δύο αριθμών είναι: {result}")
except ValueError:
    print("Παρακαλώ δώσε έγκυρους αριθμούς.")
```
